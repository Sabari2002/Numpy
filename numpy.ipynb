```python
import numpy as np
```

```python
# ndarray.ndim
arr = np.array([[3,4,6],[0,8,1]])
dim = np.ndim(arr)
print(dim)
```

```python
# ndarray.shape
shape = np.shape(arr)
print(shape)
```

```python
# ndarray.size
size = np.size(arr)
print(size)
```

```python
# ndarray.dtype
arr = np.array([[3,4,6],[0,8,1]])
print(arr.dtype)
```

```python
# ndarray.itemsize
itemsize = arr.itemsize
print(itemsize)
```

```python
arr = np.array([1,2,3,4,5])
print(arr)
print(type(arr))
```

```python
a = np.array([[1,2,3],[4,5,6]])
print(a)
```

```python
# 0D Arrays
arr = np.array(42)
print(arr)
```

```python
# 1D arrays
arr = np.array([1,2,3,4,5])
print(arr)
```

```python
# 2D arrays
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
```
